```python
x = np.random.rand(5,6)

x = x.round(3)*100

max = 0
min = 0
sum = 0
devSum = 0
for i in x:
    for j in i:
        if j >= max:
            max = j
        if j <= min:
            min = j
        sum += j
        
avg = sum/30

for i in x:
    for j in i:
        devSum += (j-avg)**2

variance = devSum/30

stdDev = variance ** 0.5

print(x)
print(max)
print(min)
print(sum)
print(avg)
print(devSum)
print(variance)
print(stdDev)

print(np.max(x), np.min(x), np.sum(x), np.average(x), np.var(x), np.std(x))
```

```python
def makeRandom(low, high, num):
    high += 1
    result = []
    while(True):
        randValue = np.random.randint(low,high)
        if randValue not in result:
            result.append(randValue)
            
        if len(result) == num:
                return result
            
            
def intersection(inA, inB):
    result = []
    
    for i in inA:
        if i in inB:
            result.append(i)
        result.sort()

    return result

print(intersection(makeRandom(1,30,10),makeRandom(1,30,10)))    
```
